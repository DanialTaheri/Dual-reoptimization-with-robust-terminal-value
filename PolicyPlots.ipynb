```python
# Finding coefficients as a function of [res, cond, capacity]
from piecewise_linear_regression.piecewise_estimator import PiecewiseRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.linear_model import LinearRegression
import numpy as np
import json
from math import sqrt
from sklearn.metrics import mean_squared_error
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
def read_table(file_name):
    with open(file_name) as json_file:
        table = json.load(json_file)
    return table
nb_plots = 6
RMSE_linear = []
NRMSE_mean_y_linear = []
NRMSE_range_linear = []
RMSE_Plinear = []
NRMSE_mean_y_Plinear = []
NRMSE_range_Plinear = []
parameters = {}
for i in range(1, nb_plots):
    Table1 = read_table('RobustTables/Table-modelrobust-category%s-capacity36000.json'%i)
    Table2 = read_table('RobustTables/Table-modelrobust-category%s-capacity42000-gran2000.json'%i)
    Table3 = read_table('RobustTables/Table-modelrobust-category%s-capacity48000-gran2000.json'%i)

    features = []
    value = []
    Month = '0'
    for nb, Table in enumerate([Table1, Table2, Table3]):
        if nb == 0:
            cap = 36000
        elif nb == 1:
            cap = 42000
        else:
            cap = 48000
        for elem in Table[Month]:
            features.append([elem[0], elem[1]/100., cap]) 
            value.append(elem[2])

    X_train, X_test, y_train, y_test = train_test_split(features, value, test_size=0.25, random_state=42)
    reg = LinearRegression()
    reg.fit(X_train, y_train)
    parameters['category%s'%i] = [{'feature coeff':list(reg.coef_), 'intercept': reg.intercept_}]
    y_pred = reg.predict(X_test)
    RMSE_linear.append(sqrt(mean_squared_error(y_test, y_pred)))
    NRMSE_mean_y_linear.append(sqrt(mean_squared_error(y_test, y_pred))/(np.mean(y_test)))
    NRMSE_range_linear.append(sqrt(mean_squared_error(y_test, y_pred))/(max(y_test) - min(y_test)))
    

    model = PiecewiseRegressor(verbose=True,
                               binner=DecisionTreeRegressor(min_samples_leaf=5000))
    model.fit(np.array(X_train), np.array(y_train))
    print('Table: {}'.format(i))
    print(model._report_coefficient())
    print("associations", model.association_rule())
    y2_pred = model.predict(np.array(X_test))
    RMSE_Plinear.append(sqrt(mean_squared_error(y_test, y2_pred)))
    NRMSE_mean_y_Plinear.append(sqrt(mean_squared_error(y_test, y2_pred))/(np.mean(y_test)))
    NRMSE_range_Plinear.append(sqrt(mean_squared_error(y_test, y2_pred))/(max(y_test) - min(y_test)))
    
plt.plot(np.arange(1, nb_plots), NRMSE_mean_y_Plinear, label="piecewise linear") 
plt.plot(np.arange(1, nb_plots), NRMSE_mean_y_linear, label="linear") 
plt.xticks(np.arange(1., nb_plots, 1.0))
plt.title('RMSE divided by mean value')
plt.xlabel("RVI table category")
plt.ylabel('Normalized RMSE')
#plt.ylim(0, 0.5)
plt.legend()
plt.show()
          
plt.plot(range(1, nb_plots), NRMSE_range_Plinear, label="piecewise linear") 
plt.plot(range(1, nb_plots), NRMSE_range_linear, label="linear") 
plt.xticks(np.arange(1., nb_plots, 1.0))
plt.title('RMSE divided by range of values')
plt.xlabel("RVI table category")
plt.ylabel('Normalized RMSE')
#plt.ylim(0, 0.5)
plt.legend()
plt.show()

plt.plot(range(1, nb_plots), RMSE_Plinear, label="piecewise linear")
plt.plot(range(1, nb_plots), RMSE_linear, label="linear")
plt.xticks(np.arange(1., nb_plots, 1.0))
plt.title('RMSE')
plt.xlabel("RVI table category")
plt.ylabel('RMSE')
#plt.ylim(0, 25e+5)
plt.legend()
plt.show()    
```
